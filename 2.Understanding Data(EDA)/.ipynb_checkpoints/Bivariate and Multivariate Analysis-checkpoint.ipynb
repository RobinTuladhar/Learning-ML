# Bivariate Analysis
Bivariate analysis is a statistical method that involves analyzing the relationship between two variables. In other words, it is the analysis of two variables simultaneously, to determine whether there is any association or correlation between them.

In bivariate analysis, we typically use techniques such as correlation analysis, regression analysis, and contingency table analysis to examine the relationship between the two variables.

Bivariate analysis can be useful for identifying patterns and relationships between variables, and for predicting the values of one variable based on the values of the other variable. It can also help to identify any confounding variables that may be influencing the relationship between the two variables.

Bivariate analysis is an important tool in many areas of research, including social sciences, business, and healthcare. It can help researchers to understand the complex relationships between different factors and to develop more accurate models for predicting outcomes.

# Multivariate Analysis
Multivariate analysis is a statistical method that involves analyzing more than two variables simultaneously. It is the study of the relationship between multiple variables, and how they influence each other.

In multivariate analysis, we use techniques such as multivariate regression analysis, factor analysis, and principal component analysis to examine the relationships between multiple variables.

Multivariate analysis can be useful for identifying complex patterns and relationships between multiple variables. It can help researchers to identify which variables are most strongly related to a particular outcome, and to develop more accurate models for predicting outcomes.

Multivariate analysis is commonly used in fields such as finance, marketing, social sciences, and healthcare. It can help researchers to understand the complex relationships between different factors and to develop more effective strategies for achieving desired outcomes.

# 1.Satterplot (Numerical - Numerical)
Generally gives a linear corelation

```python
# There is a linear relation between bill and the amount of bill
sns.scatterplot(tips["total_bill"], tips["tip"] ) #bivariate
```

```python
sns.scatterplot(tips["total_bill"], tips["tip"], hue = tips["sex"]) #multivariate
```

```python
# this plot shows relation between 4 columns. 
sns.scatterplot(tips["total_bill"],tips["tip"], hue = tips["sex"], style = tips["smoker"],)
```

```python
# this plot shows relation between 5 columns. 
sns.scatterplot(tips["total_bill"],tips["tip"], hue = tips["sex"], style = tips["smoker"], size = tips["size"])
```
