# Decision Tree
A decision tree is a popular machine learning algorithm used for both regression and classification tasks. It is a type of model that works by partitioning the data into subsets based on different attributes, with the aim of creating a tree-like structure that can be used to make predictions.

The decision tree algorithm is used because it is easy to understand and interpret, can handle both numerical and categorical data, and is relatively fast to train on large datasets. It is also a non-parametric algorithm, which means that it does not require any assumptions about the underlying distribution of the data.

### Some of the benefits of using decision trees include:

- Easy to interpret and explain: Decision trees can be visualized and easily explained, making them accessible to a wide range of users.

- Can handle both numerical and categorical data: Decision trees can handle a mix of data types, making them useful for a variety of applications.

- Can handle missing data: Decision trees can handle missing data, which is useful in real-world scenarios where data is often incomplete.

### However, decision trees also have some limitations:

- Can overfit the data: Decision trees can create overly complex models that fit the training data too closely, leading to poor generalization on new data.

- Can be sensitive to small changes in the data: Decision trees can be sensitive to small changes in the data, which can lead to different tree structures and predictions.

- Can be biased towards features with many levels: Decision trees can be biased towards features with many levels, which can lead to over-representation of these features in the tree.

# Entropy
Entropy is a concept used in decision trees to measure the impurity of a set of data. In the context of decision trees, entropy represents the degree of disorder or randomness in the data.

More formally, entropy is defined as the sum of the negative logarithm of probabilities of all possible outcomes, weighted by their probabilities. In a decision tree, entropy is used to calculate the information gain of splitting the data based on a particular attribute. The information gain is the reduction in entropy achieved by partitioning the data based on the attribute, and it is used to determine the best attribute to split on at each node of the tree.

Intuitively, if a set of data is completely pure (i.e., all samples belong to the same class), then the entropy is 0. On the other hand, if the data is completely impure (i.e., the samples are evenly distributed across all classes), then the entropy is 1.

In summary, entropy is a measure of impurity in decision trees that is used to select the best attribute to split on at each node of the tree. It is calculated as the sum of the negative logarithm of probabilities of all possible outcomes, weighted by their probabilities.



# Information Gain
Information gain is a concept used in decision trees to measure the effectiveness of splitting a set of data based on a particular attribute. The information gain of an attribute is the difference between the entropy of the parent node (i.e., the set of data before the split) and the weighted average of the entropies of the child nodes (i.e., the sets of data after the split).

In other words, information gain measures how much the entropy of the data is reduced by splitting it based on a particular attribute. The higher the information gain, the more effective the attribute is at reducing the entropy and separating the data into homogeneous groups with respect to the target variable.

Information gain is used in decision tree algorithms to determine the best attribute to split on at each node of the tree. The attribute with the highest information gain is chosen as the splitting criterion, as it is expected to result in the greatest reduction of entropy and the most accurate predictions.

In summary, information gain is a measure of the effectiveness of splitting a set of data based on a particular attribute in a decision tree. It is calculated as the difference between the entropy of the parent node and the weighted average of the entropies of the child nodes, and it is used to determine the best attribute to split on at each node of the tree.

## Visualizing the Decision Tree

```python
from sklearn.datasets import load_iris # sklearn has some datasets
```

```python
df = load_iris()
```

```python
X = df.data # independent variables
y = df.target # dependent variables
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 12) # note that accurcay may vary at diff random_states
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import plot_tree
from matplotlib.pylab import rcParams
rcParams["figure.figsize"] = 80,50
plot_tree(dt)
```
