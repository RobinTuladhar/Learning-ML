```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head(10)
```

# Working on number column

```python
df["number"].unique()
```

```python
fig = df["number"].value_counts().plot.bar()
fig.set_title("Passengers travelling with")
```

```python
# Extracting numbercial part
df["number_numeric"] = pd.to_numeric(df["number"], errors="coerce", downcast="integer")
df["number_numeric"]
```

```python
#Extracting categorical part
df["number_categorical"] = np.where(df["number_numeric"].isnull(), df["number"], np.nan)
df["number_categorical"]
```

```python
df.head()
```

# Working with Cabin column

```python
df["Cabin"].unique()
```

```python
df["Ticket"].unique()
```

```python
df["cabin_cat"] = df["Cabin"].str[0] # captures the first letter
df["cabin_num"] = df["Cabin"].str.extract('(\d+)') # extracts numberical part
df.head()
```

```python
df['cabin_cat'].value_counts()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
