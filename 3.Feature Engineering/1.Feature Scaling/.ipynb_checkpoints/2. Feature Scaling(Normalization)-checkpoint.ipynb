# Normalization 
is the process of organizing data in a database to reduce redundancy and dependency. It involves breaking down a database into smaller, more manageable tables and establishing relationships between them. The goal of normalization is to eliminate data redundancy, which can lead to inconsistencies and errors in data.

There are several normal forms in database normalization, each building on the previous form and providing greater levels of data consistency and accuracy. The most commonly used normal forms are:

First normal form (1NF): Ensures that each table has a primary key and that each column in the table contains atomic values (i.e., values that cannot be broken down further).

Second normal form (2NF): Builds on 1NF and ensures that each non-primary key column in a table is dependent on the table's primary key.

Third normal form (3NF): Builds on 2NF and ensures that each non-primary key column in a table is independent of other non-primary key columns in the same table.

There are higher normal forms as well, such as Boyce-Codd normal form (BCNF) and fourth normal form (4NF), but they are less commonly used.

Normalization is an important part of database design as it helps to improve data consistency, reduce data redundancy, and make it easier to maintain and modify a database.

## Types
### 1. Min-max Scaling:

This technique scales the features to a range between 0 and 1. It is calculated as (x - min) / (max - min), where x is the feature value, min is the minimum feature value, and max is the maximum feature value. Min-max normalization is useful for features that have a limited range and for algorithms that are sensitive to differences in scale.

### 2. Mean Normalization:

Mean normalization is a specific type of normalization technique that involves scaling the features so that they have a mean of zero. This technique is useful for features that have different means and for algorithms that are sensitive to differences in mean values.

To perform mean normalization, the following steps are typically taken:

Calculate the mean of the feature values.
Subtract the mean from each feature value.
Scale the features by dividing each feature value by the range of the feature values.
The scaled feature value is calculated as (x - mean) / (max - min), where x is the feature value, mean is the mean of the feature values, max is the maximum feature value, and min is the minimum feature value.

Mean normalization can be useful for machine learning algorithms that are sensitive to differences in scale and mean values, such as clustering algorithms and some types of neural networks. However, it may not be appropriate for all types of data or machine learning problems, and other types of normalization or scaling techniques may be more appropriate.


### 3. Max-Absolute:
Max-absolute scaling is a specific type of feature scaling technique that involves scaling the features by dividing each feature value by the maximum absolute value of that feature across all data points. This technique is useful for features that have a large range and for algorithms that are sensitive to differences in scale.

To perform max-absolute scaling, the following steps are typically taken:

Find the maximum absolute value of each feature across all data points.
Divide each feature value by the corresponding maximum absolute value.
The scaled feature value is calculated as x / max(abs(x)), where x is the feature value.

Max-absolute scaling can be useful for machine learning algorithms that are sensitive to differences in scale, such as support vector machines (SVMs) and deep neural networks. It is particularly useful when the data contains extreme values, as it ensures that these values do not dominate the other feature values in the analysis. However, it may not be appropriate for all types of data or machine learning problems, and other types of scaling or normalization techniques may be more appropriate.

### 4. Robust-Scaling:
Robust scaling is a specific type of feature scaling technique that is designed to be more robust to outliers in the data. It involves scaling the features to a range between 0 and 1 based on their interquartile range (IQR) instead of their minimum and maximum values, which can be sensitive to outliers.

To perform robust scaling, the following steps are typically taken:

Calculate the median of each feature.
Calculate the interquartile range (IQR) of each feature as the difference between the 75th percentile and the 25th percentile.
Divide each feature value by the corresponding IQR value.
The scaled feature value is calculated as (x - median) / IQR, where x is the feature value, median is the median of the feature values, and IQR is the interquartile range of the feature values.

Robust scaling can be useful for machine learning algorithms that are sensitive to outliers, such as clustering algorithms and some types of regression models. By scaling the features based on their IQR instead of their minimum and maximum values, this technique reduces the impact of outliers on the feature scaling process. However, it may not be appropriate for all types of data or machine learning problems, and other types of scaling or normalization techniques may be more appropriate.

# MIN-MAX SCALING

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2]) # Using only 3 columns from the entire dataset
df
```

```python
df.columns=['Class label', 'Alcohol', 'Malic acid']
df
```

```python
plt.figure(figsize= (10,10))
plt.legend("Alcohol")
sns.kdeplot(df['Alcohol']) # Showing the bi-modle distribution
```

```python
plt.figure(figsize= (10,10))
sns.kdeplot(df['Malic acid'])
```

```python
color_dict={1:'red',3:'green',2:'blue'}
sns.scatterplot(df['Alcohol'],df['Malic acid'],hue=df['Class label'],palette=color_dict)
```

# Remember before scaling train_test_split is a must

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label', axis=1), df['Class label'], test_size=0.3, random_state=0)
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
# fit only the training data
scalar.fit(X_train)
# transform both trainign and testing data
X_train_scaled = scalar.transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
# Whenever scikit learn's classes is used it converts the result to numpy array so to handle those
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
# Describing the old data
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1) # It can be seen that the min=0 and the max=1
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Alcohol'], X_train['Malic acid'],c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic acid'],c=y_train)
ax2.set_title("After Scaling")
plt.show()
## THe data in the below figure is complete same but check the scale on both the axis it is scaled to one meaning the dataset has been squeezed to a rectangle ranging 0-1 in 
## both the axis
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alcohol'], ax=ax1)
sns.kdeplot(X_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
sns.kdeplot(X_train_scaled['Malic acid'], ax=ax2)
plt.show()
# Check in the first figure
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Alcohol Distribution Before Scaling')
sns.kdeplot(X_train['Alcohol'], ax=ax1)

# after scaling
ax2.set_title('Alcohol Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Malic acid Distribution Before Scaling')
sns.kdeplot(X_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('Malic acid Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Malic acid'], ax=ax2)
plt.show()
```
