# Logistic Regression
is like a simple but effective tool in the world of machine learning, especially when you want to predict whether something belongs to one of two groups, like "yes" or "no," "spam" or "not spam." It's a fundamental algorithm known for its simplicity and interpretability.Logistic regression uses a sigmoid function to transform the linear combination of the input features into a probability. The sigmoid function is a non-linear function that outputs a value between 0 and 1. This ensures that the predicted probability is always between 0 and 1, which makes sense for a binary outcome.

Imagine you are trying to build a model to predict whether a customer will churn (cancel their subscription) or not. You have a dataset of customers with their churn status and other features, such as their age, gender, and subscription plan.

You could start by building a simple logistic regression model with the following features:

age
gender
subscription_plan

The logistic regression model would then learn the relationship between these features and the churn status. Once the model is trained, you could use it to predict the probability of churn for a new customer.

For example, if the new customer is a young male with a monthly subscription plan, the model might predict that the customer has a 20% chance of churning.


## Objective: 

Imagine you have a magical ruler that measures the height of students and tells you whether they're likely to pass an exam (yes) or fail (no). Logistic Regression is like using this ruler to make predictions based on a student's height.

## Explanation:

1. Binary Classification: Logistic Regression is primarily used for binary classification problems, where the goal is to predict one of two possible outcomes. For example, it can predict whether an email is spam (yes) or not spam (no) based on various features.

2. S-Shaped Curve: The algorithm uses a mathematical function called the logistic function or sigmoid function, which creates an S-shaped curve. This curve maps input features to a probability score between 0 and 1, indicating the likelihood of belonging to one of the two classes.

3. Interpretable Coefficients: Logistic Regression provides interpretable coefficients for each input feature. These coefficients tell you how much each feature influences the prediction. Positive coefficients increase the probability of the "yes" class, while negative coefficients decrease it.


## Advantages:

1. Interpretability: Logistic Regression is easy to interpret. You can understand how each feature affects the prediction by looking at the coefficients.
2. Simple and Fast: It's computationally efficient and works well with smaller datasets.
3. Works with Few Features: It's effective when you have a small number of input features.
4. Easy to understand and interpret: Logistic regression models are relatively easy to understand and interpret, even for beginners. This is because logistic regression models are linear in the parameters.
5. Robust to outliers: Logistic regression models are robust to outliers, which means that they are not overly sensitive to individual data points that may be different from the rest of the data.
6. Efficient to train: Logistic regression models can be trained efficiently, even on large datasets. This is because there are many efficient algorithms available for training logistic regression models.

## Disadvantages:

1. Limited to Linear Relationships: Logistic Regression assumes that the relationship between features and the log-odds of the response variable is linear. It may not work well for complex, nonlinear relationships.
2. Binary Classification Only: Logistic Regression is designed for binary classification. Extending it to multi-class problems requires special techniques.
3. Sensitive to Outliers: Outliers in the data can have a significant impact on Logistic Regression.
4. Can be prone to overfitting: Logistic regression models can be prone to overfitting, which means that they can learn the training data too well and not generalize well to new data. This can be mitigated by using regularization techniques, such as L1 or L2 regularization.
5. Can be biased: Logistic regression models can be biased, depending on the way the training data is collected. For example, if the training data is not representative of the population that the model will be used to make predictions on, then the model will be biased.

## Common Applications:

1. Medical Diagnosis: It's used in healthcare for predicting the presence of a disease (e.g., diabetes) based on patient data.
2. Credit Scoring: Banks and financial institutions use it to assess the creditworthiness of applicants.
3. Marketing: In marketing, it helps predict customer churn (whether a customer will leave) or click-through rates for online ads.
4. Spam Detection: Logistic Regression is commonly used in spam email detection.

In summary, Logistic Regression is a simple yet effective tool for binary classification problems, offering interpretability and ease of use. It's valuable in scenarios where you want to predict one of two outcomes based on input features, making it a fundamental technique in machine learning.

# Logistic Regression from scratch using Gradient Descent

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(penalty='none',solver='sag')
lor.fit(X,y)
```

```python
print(lor.coef_)
print(lor.intercept_)
```

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b1 = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m1*x_input + b1
```

```python
def gd(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    
    for i in range(5000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])
        
    return weights[1:],weights[0]
        
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python

coef_,intercept_ = gd(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
